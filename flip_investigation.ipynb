```python
import SimpleITK as sitk
import numpy as np
import matplotlib.pyplot as plt
import preprocessing as pre
import os
plt.rcParams['image.cmap'] = 'gray'
```

```python
os.listdir('/Volumes/SAMSUNG/visceral_full')

def display_scan(scan, sl, side='c'):
    plt.figure(figsize=(10, 10))
    if side == 'c':
        plt.imshow(scan[:, sl, :])
    elif side == 'a':
        plt.imshow(scan[:, :, sl])
    else:
        plt.imshow(scan[sl, :, :])
    plt.show()
```

```python
img = pre.get_image_given_path('/Volumes/SAMSUNG/visceral_full/test/10000005_1_CT_wb.nii.gz')
print(img.GetSize()) # s, c, a

img_arr = sitk.GetArrayFromImage(img)
print(img_arr.shape) # a, c, s
```

```python
display_scan(img_arr, 173, 'c')
display_scan(img_arr, 229, 'a') # switched with s
display_scan(img_arr, 572, 's') # switched with a
```

```python
img_arr_t = img_arr.T
# image is still not right on the coronal / y axis, flip around z axis
img_arr_t = img_arr_t[:, :, ::-1]
print(img_arr_t.shape) # s, c, a
```

```python
display_scan(img_arr_t, 205, 'c') # 
display_scan(img_arr_t, 308, 'a')
display_scan(img_arr_t, 229, 's')
```
